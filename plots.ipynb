```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("yaxis")
plt.show()
```

```python
x=[10,20,30,40]
plt.pie(x)
plt.show()
mylabels=("Apples","banannas","cheeries","dates")
plt.pie(x,labels=mylabels)
plt.show()
```

```python
y=(15,25,35,45)
mylabels=("Apples","banannas","cheeries","dates")
mycolors=("black","hotpink","blue","orange")
plt.pie(y,labels=mylabels , colors=mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt
y=[15,25,35,45]
mylabels=["Apples","banannas","cheeries","dates"]
myexplode=[0.2,0,0,0.5]
plt.pie(y, labels=mylabels , explode=myexplode , shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[3,6,8,5]
y=[3,7,8,9]
plt.title("barplot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.bar(x,y)
```

```python
import matplotlib.pyplot as plt
x=[3,6,8,5]
y=[3,7,8,9]
plt.title("barplot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(x,y)
```

```python
x=[1,3,5,7]
y=[2,4,6,7]
colors=["g","r","b","orange"]
sizes=[500,600,500,500]
plt.scatter(x,y,c=colors ,s=sizes ,alpha=1, marker="*",edgecolor="r")
plt.title("scatter plot",fontsize=15)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.scatter(x,y)
plt.show()
```
