```python
import pandas  as pd
import matplotlib.pyplot as plt
```

```python
data={"date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2023-03-01","2023-04-01"]),
                            "value":[5,10,20,30,15,12,20,30]}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("date")
plt.ylabel("value")
plt.title("Time series with Seasonality ")
plt.grid(True)
plt.show()
```
